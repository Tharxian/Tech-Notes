# Matplotlib的使用
Matplotlib为效仿MATLAB的一个图表绘制的模组

### 绘制随机变量数组的图像
$x,y$ 分别收集了100个标准正态分布的随机变量，有Mean: $\mu = 0$, standard division 是 1

```python
from matplotlib import pyplot as plt
import numpy as np

# 生成随机变量数组
x = np.random.normal(0,1,100)
y =  np.random.normal(0,1,100)

# 打印它们的数据维度
print(x.shape,y.shape)
#绘制图表
plt.scatter(x,y)
# 展示图表
plt.show()
```

- 图表绘制的颜色会默认是蓝色，更改颜色可以使用 `plt.scatter(x,y,color='r')`

- 图表绘制的图形会默认是小圆点，更改图形可以使用 `plt.scatter(x,y,maker=x')`

### 同一图像下绘制多组数据


```python
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)
plt.scatter(x,y,color = 'r',marker='x')

m = np.random.normal(0.5,1,100)
n = np.random.normal(0.5,1,100)
plt.scatter(m,n,color = 'g')

plt.show()
```
